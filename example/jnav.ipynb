```python
import requests
```

```python
def download_image_from_url(url: str, filename: str):
    # url="https://images.deliveryhero.io/image/fd-th/Products/3678508.png"
    r = requests.get(url)
    if r.ok:
        print(r.status_code)
        content = r.content  # binary data
        with open(filename, mode="wb") as f:  # wb = [w]rite [b]inary data to file
            f.write(content)
    else:
        print(r.status_code)
        print("error...")

```

```python
for i, p in enumerate(product_lists, start=1):
    print(f"{i}: {p['name']}: {p['images'][0]['image_url']}")
    menu = p["name"] + ".png"
    download_image_from_url(p["images"][0]["image_url"], menu)

```
