```python
def rectangle(w, h):
	"""calculate rectangular area
	Parameters:
		w (_type_): width
		h (_type_): สูง

	Returns:
		_type_: พื้นที่สี่เหลี่ยมผืนผ้า
	"""
	return w * h
```

```python
def triangle(b:float, h:int)->float:
	"""_summary_

	Args:
		b (float): _description_
		h (int): _description_

	Returns:
		float: _description_

	Examples:
	>>> triangle(5, 10)
	>>> triangle(2, 7)

	Tests:
	>>> triangle(5, 10)
	25.0
	>>> triangle(2, 7)
	7.0
	"""
	return .5 * b * h
```

```python
triangle(5, 10)
```

```python
triangle(5, 7)
```
