# f-string cheat sheet
- ref: https://docs.python.org/3/tutorial/inputoutput.html

```python
def demo2():
    mi = 7
    print(f"{mi:03}")
    amt = 12750.7534
    print("amount = {:,.3f}".format(amt))
    print(f"amount = {amt}")
    print(f"amount = {amt:.2f}")
    print(f"amount = {amt:,.2f}")
    print(f"{amt:,.2f}".rjust(15, "+"))
    print(f"{amt:,.2f}".rjust(15, "-"))
    print(f"{amt:,.2f}".ljust(15, "-"))
    print(f"{amt:,.2f}".center(15, "-"))


demo2()
```

```python
def demo3():
    w = 70.5
    h = 1.72
    print(f"weight = {w:.2f}kg., height = {h:.1f}m., BMI = {w / (h * h):.2f}")


demo3()
```

```python
def demo4():
    v = 4672.25
    print(f":{v:>20}:")
    print(f":{v:*>20}:")
    print(f":{v:->20}:")
    print(f":{v:+>20}:")
    print(f":{v:<20}:")
    print(f":{v:-<20}:")
    print(f":{v:20}:")
    print(f":{v:^20}:")


demo4()
```

```python
def demo5():
    v = "Peter"
    print(f":{v:>20}:")
    print(f":{v:*>20}:")
    print(f":{v:->20}:")
    print(f":{v:+>20}:")
    print(f":{v:<20}:")
    print(f":{v:-<20}:")
    print(f":{v:20}:")
    print(f":{v:^20}:")


demo5()
```

```python
def demo_percent():
    p = 0.125
    print(f"{p:.2%}")


demo_percent()
```

```python
def demo_tuple():
    point = (4, 10)
    print(f"({point[0]}, {point[1]})")


demo_tuple()
```
