```python
# destination = "วัดเจ้ามูล"
destination = "สำนักงานเขตบางกอกใหญ่"


def station(destination: str, stations: dict) -> str:
    """หาว่าปลายทางอยู่สถานีใด ทางออกใด

    Args:
        destination (str): ปลายทางที่จะไป เช่น วัดเจ้ามูล
        stations (dict): ข้อมูลสถานีและทางออกรถไฟฟ้า MRT

    Returns:
        str: สถานีปลายทางและหมายเลขประตูทางออก

    Examples:
    >>> station(destination, d)
    'BL01 (ท่าพระ) -> exit 2B'
    """
    for mrt in stations.keys():
        # print(mrt)
        for place in stations[mrt]["exit"]:
            if destination in stations[mrt]["exit"][place]:
                return f'{mrt} ({stations[mrt]["name"]["th"]}) -> exit {place}'
                # print(f'{mrt} ({d[mrt]["name"]["th"]}) -> exit {place}')


station(destination, stations)
```
