```python
import matplotlib.pyplot as plt
import numpy as np
import collections
```

```python

def demo1():
    x = range(4)
    y = (20, 25, 40, 30)
    avg = sum(y) / 4
    print(avg)
    xtick = ("mocha", "latte", "espresso", "tea")
    plt.bar(x, y, color="green")
    plt.xticks(x, xtick)
    plt.axhline(avg, color="red", linestyle="--")
    plt.title("Orders by menu\nFeb 2017", color="orange", fontsize=17)
    plt.ylabel("# orders")
    plt.show()

demo1()
```

```python

def bar_text_freq():
    flowers = ["mocha", "latte", "espresso", "tea", "water"]
    prob = [.05, .25, .20, .1, .4]
    d = np.random.choice(flowers, 50, p=prob)
    print(d)
    f = collections.Counter(d)
    print(list(f.items()))
    x = range(len(flowers))
    plt.bar(x, f.values(), edgecolor="black")
    plt.xticks(x, f.keys())
    plt.show()

bar_text_freq()
```

```python
def barv():
    x = range(4)
    y = (20, 25, 40, 30)
    xticks = ("mocha", "latte", "espresso", "tea")
    plt.xticks(x, xticks)
    plt.bar(x, y)
    plt.show()

barv()
```

```python

def barh():
    x = range(4)
    y = (20, 25, 40, 30)
    xticks = ("mocha", "latte", "espresso", "tea")
    plt.yticks(x, xticks)
    plt.barh(x, y)
    plt.show()

barh()
```

```python

def bar_subplot():
    x = range(4)
    y = (20, 25, 40, 30)
    xticks = ("mocha", "latte", "espresso", "tea")
    fig, ax = plt.subplots(2, 1)
    ax[0].bar(x, y, color = "green")
    ax[1].barh(x, y, color = "purple")
    plt.sca(ax[0])
    plt.xticks(x, xticks)
    plt.sca(ax[1])
    plt.yticks(x, xticks)
    fig.tight_layout()
    plt.show()

bar_subplot()
```
