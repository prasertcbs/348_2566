````python
import numpy as np
import plotly.express as px
import random


def heart():
    # http://mathworld.wolfram.com/HeartCurve.html
    t = np.linspace(0, 2 * np.pi, 360)
    # print(t)
    x = 16 * np.sin(t) ** 3
    y = 13 * np.cos(t) - 5 * np.cos(2 * t) - 2 * np.cos(3 * t) - np.cos(4 * t)
    # return px.scatter(x=x, y=y)
    fig = px.scatter(x=x, y=y, width=900, height=900)
    for _ in range(10):
        fig.add_scatter(
            x=(x + random.randrange(5, 30)) * random.randrange(10, 200) / 100,
            y=(y + random.randrange(5, 30)) * random.randrange(10, 200) / 100,
        )
        # fig.add_scatter(x=x + 20, y=y)
    fig.update_layout(showlegend=False)
    return fig


heart()
````
