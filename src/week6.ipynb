```python
def discount_price(amt, member_level):
    discount_pct = 0
    if member_level=='bronze':
        discount_pct = .05
    elif member_level=='silver':
        discount_pct = .10
    elif member_level=='gold':
        discount_pct = .20
    return amt * (1-discount_pct), discount_pct

amt = 100
v=discount_price(amt, 'silver')
v
# for member_card in ['bronze', 'silver', 'gold']:
#     print(f"before discount = {amt}, member_card = {member_card} {discount_price(amt, member_card)}")
```

```python
def split_fullname(fullname):
    names=fullname.split(' ')
    return names

def get_firstname(fullname):
    names=fullname.split(' ')
    return names[0]

def get_lastname(fullname):
    names=fullname.split(' ')
    return names[1]

if __name__ == '__main__':
    fullname='Tony Stark'
    print(get_firstname(fullname))
```

```python
s=split_fullname('Peter Parker')
print(s[0])
print(s[1])
```
