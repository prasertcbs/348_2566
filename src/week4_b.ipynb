# for loop

```python
# 2 + 4 + 6 + 8 + ... + 100 = ?
total = 0
for i in range(0, 101, 2):
    # print(i)
    total = total + i
print(total)
```

# while loop

```python
total = 0
i = 0
while i < 101:
    # print(i)
    total = total + i # total += i
    i += 2 # i = i + 2
print(total)
```
