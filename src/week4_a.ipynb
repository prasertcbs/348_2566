# for loop
- บวกเลขคู่ตั้งแต่ 0-100 พร้อมแสดงผล

```python
total = 0
for i in range(0, 101, 2):
    print(i, end=" ")
    total = total + i
print()
print(f'total = {total}')
```

# while loop

```python
total = 0
i = 0
while i < 101:
    print(i, end=" ")
    total = total + i
    i += 2
print()
print(f'total = {total}')
```

# lucky car plate

```python
def sum_digit(n):
	total = 0
	vals=list(str(n))
	print(f'vals = {vals}', end=" ")
	for v in vals:
		total += int(v)
	return total

print(sum_digit(1235))
```

```python
for i in range(1234, 2000):
    print(sum_digit(i))
```
