# Thai National ID check digit

![](https://wikimedia.org/api/rest_v1/media/math/render/svg/d77d4d118abb6f73a05cb882d5a12a8b27c55c90)

```python
nid='1203289712342'
seq=13
total=0
for c in nid[:12]:
    n=int(c)*seq
    total += n
    seq -= 1
x = total % 11
if x <= 1:
    check_digit=1-x
else:
    check_digit=11-x
print(check_digit)
```

```python
nid='1203289712342'
total=0
for seq, c in zip(range(13, 1, -1), nid[:12]):
    total += int(c)*seq
x = total % 11
check_digit=1-x if x <= 1 else 11-x
print(check_digit)
```

```python
nid='1203289712342'
def thai_nid_check_digit(nid):
    x=sum([int(c)*seq for seq, c in zip(range(13, 1, -1), nid[:12])]) % 11
    return 1-x if x <= 1 else 11-x
```
